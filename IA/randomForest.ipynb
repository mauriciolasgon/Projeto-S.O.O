```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import mean_squared_error
```

```python
arquivo="data\dataset.csv"
dataset=pd.read_csv(arquivo)
dataset.head(10)

symptoms="data\Symptom-severity.csv"
symptoms=pd.read_csv(symptoms)

data=pd.DataFrame(index=dataset["Disease"],columns=symptoms['Symptom'])


data


```

```python

dataset.head(20)
```

```python



x=dataset.drop("Disease",axis=1)

linha_data,coluna_data=data.shape

linha_x,coluna_x=x.shape

for sintoma,j in zip(data,range(coluna_data)):
    for sintomas,i in zip(x.iterrows(),range(linha_x)):
        if(sintomas[1].str.contains(sintoma,case=False).any()):
            data.iloc[i][j]=1
        else:
            data.iloc[i][j]=0





data.head(20)
```

```python

y=dataset["Disease"]

le=preprocessing.LabelEncoder()

#for colum in x.columns:
    #x[colum] = le.fit_transform(x[colum])

y=le.fit_transform(y)
```

```python
Xtr,Xval,Ytr,Yval=train_test_split(data,y,test_size=0.6,random_state=0)

floresta=RandomForestRegressor(n_estimators=300,random_state=0)

floresta.fit(Xtr,Ytr)


p=floresta.predict(Xval)


np.sqrt(mean_squared_error(Yval,p))

```
